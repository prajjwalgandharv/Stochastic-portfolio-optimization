```python
import os
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
try:
    os.mkdir("out")
    print("directory created")
except FileExistsError:
    print("out folder already exists")
except OSError as e:
    print(f"error creating out folder: {e}")
```

```python
tickers_csv = os.path.join("out", "tickers.csv")
tickers = ["AAPL","MSFT","GOOGL","AMZN","META","NVDA","AVGO","ORCL","IBM","SAP",
        "JPM","BAC","WFC","MS","GS","BLK","V","MA","BRK-B","KO",
        "PG","PEP","WMT","COST","HD","MCD","DIS","NKE","GE","UNH",
        "LLY","PFE","MRK","ABT","TMO","AMGN","ADBE","CRM","NOW","SNOW",
        "AMD","TSM","ASML","AMAT","TSLA","FSLR","ENPH","NEE","PLTR","PANW"]
pd.DataFrame({"ticker": tickers}).to_csv(tickers_csv, index=False) 
```

```python
# Date range for collecting stock price data (including the initial quarter lookback before portfolio creation)
# starts collecting data from April 1, 2024 with the period from April 1, 2024 to June 30, 2024 being used as the lookback
# to help portfolio allocation on July 1, 2024. Stays updated to the latest stock movements
start = "2024-04-01"
end = None
```

```python
data = yf.download(tickers, start = start, end = end, auto_adjust = True)
```

```python
prices = data["Close"]
```

```python
prices = prices.dropna(how = "all").sort_index()
prices = prices.dropna(axis = 1, how = "any")
```

```python
pct_returns = prices.pct_change().dropna()
```

```python
prices
```

```python
prices_out = prices.reset_index().rename(columns = {"Date": "Date"})
returns_out = pct_returns.reset_index().rename(columns = {"Date": "Date"})
```

```python
prices_out
```

```python
prices_out.to_csv("out/prices.csv", index=False)
returns_out.to_csv("out/returns.csv", index=False)
```
